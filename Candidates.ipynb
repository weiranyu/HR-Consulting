```python
## Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Reading data
dfActivities = pd.read_csv('activities.csv')
dfCandidates = pd.read_csv('candidateprogress.csv')
```

```python
## Fixing date
dfCandidates['Date'] =  pd.to_datetime(dfCandidates['Date'].astype(str), format='%Y%m%d')
dfCandidates.head(5)
```

```python
dfCandidates.groupby(by=['Candidate Name', 'Candidate ID']).max()
```

```python
dfActivities.head(5)
```

```python
## Long Format
for cols in list(dfActivities['Type of Activity'].unique()):
    dfActivities[cols] = np.where(dfActivities['Type of Activity']==cols, 1, 0)
    
dfActivities.head(5)
```

```python
## Sum on Candidate
dfCandidatesActivity = dfActivities.groupby(by=['Candidate Name', 'Candidate ID']).sum()
dfCandidatesActivity.head(5)
```

```python
## Add count of requisitions to it
dfCandidatesActivity['Requisitions Count'] = dfActivities.groupby(by=['Candidate Name', 'Candidate ID']).count()['Requisition']
dfCandidatesActivity.head(5)
```

```python
## Long Candidates Form
for cols in list(dfCandidates['Progress Item'].unique()):
    dfCandidates[cols] = np.where(dfCandidates['Progress Item']==cols, 1, 0)
    
dfCandidates.head(5)
```

```python
## Minimum & maximum Date
dfCandidatesActivity['First_Comm'] = dfCandidates.groupby(by=['Candidate Name', 'Candidate ID']).min()['Date']
dfCandidatesActivity['Last_Comm'] = dfCandidates.groupby(by=['Candidate Name', 'Candidate ID']).max()['Date']

dfCandidatesActivity.head(5)
```

```python
## Number of Days
dfCandidatesActivity['Days'] = dfCandidatesActivity['Last_Comm'] - dfCandidatesActivity['First_Comm']
dfCandidatesActivity['Days'] = (dfCandidatesActivity['Days']/np.timedelta64(1, 'D')).astype(int)
dfCandidatesActivity.head(5)
```

```python
## Sum of dfCandidates
dfCandidatesSum = dfCandidates.groupby(by=['Candidate Name', 'Candidate ID']).sum()
dfCandidatesSum.head(5)
```

```python
## Combine the 2
dfCandidatesActivity = pd.merge(dfCandidatesActivity, dfCandidatesSum, on=['Candidate Name', 'Candidate ID'])
dfCandidatesActivity.head(5)
```

```python
## Cleaning up the Pipelined and submitted columns
dfCandidatesActivity['Pipelined'] = dfCandidatesActivity[['Pipelined_x', 'Pipelined_y']].max(axis=1)
dfCandidatesActivity['Submitted'] = dfCandidatesActivity[['Submitted_x', 'Submitted_y']].max(axis=1)

## Drop the unnecessary ones
dfCandidatesActivity.drop(['Pipelined_x', 'Pipelined_y', 'Submitted_x', 'Submitted_y'], axis=1, inplace=True)
dfCandidatesActivity.head(5)
```

```python
## Export
dfCandidatesActivity.to_csv('CandidatesMaster.csv')
```
