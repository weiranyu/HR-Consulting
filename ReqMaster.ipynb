```python
## Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Reading data
dfRequisition = pd.read_csv('Application_timeline.csv')
dfRequisition.head(5)
```

```python
## Date Time
dfRequisition['First Interview Date']= pd.to_datetime(dfRequisition['First Interview Date'])
dfRequisition.dtypes
```

```python
## Reading more data
dfRequisition_Raw = pd.read_csv('requisition.csv')
dfRequisition_Raw.head(5)
```

## Cleaning dfRequisition_Raw

<font color = 'blue'> More details see the other file. Cleaning work done here is only for analysis</font>

```python
## Changing time to date-time format
dfRequisition_Raw['Date Created'] = pd.to_datetime(dfRequisition_Raw['Date Created'])
dfRequisition_Raw['Hired Date'] = pd.to_datetime(dfRequisition_Raw['Hired Date'])
dfRequisition_Raw.dtypes
```

```python
## Drop null
dfRequisition_Raw = dfRequisition_Raw.dropna(axis=0, subset=['Date Created'])
pd.isnull(dfRequisition_Raw).sum()
```

```python
## Adding a time spent finding a job column
dfRequisition_Raw['Time Spent Filling'] = dfRequisition_Raw['Hired Date'] - dfRequisition_Raw['Date Created']
dfRequisition_Raw.head(5)
```

```python
## Converting it to time
## dfRequisition_Raw['Time Spent Filling'] = (dfRequisition_Raw['Time Spent Filling']/np.timedelta64(1, 'D')).astype(int)
```

## Looking at multiple hires for the same requisition

```python
dfRequisition_Raw_Count = dfRequisition_Raw.groupby(by='Requisition').count().sort_values(by='Date Created', ascending=False)
dfRequisition_Raw_Count.head(5)
```

<font color ='blue'> We need to make a decision here. For our dataset to work, we require one date only. I suggest we only use the 'First Hired Date'</font>

```python
## Getting minimum date
dfRequisition_Raw_Min = dfRequisition_Raw.groupby(by='Requisition').min()
dfRequisition_Raw_Min.head(5)
```

## Combine the 2 dataframes

```python
dfRequisition = dfRequisition_Raw_Min.merge(dfRequisition, how='inner', left_on=['Requisition'], right_on=['Requisition'])
dfRequisition.head(5)
```

```python
## Seeing the column names
dfRequisition.columns
```

```python
## Changing the column names
dfRequisition.rename(columns={'Hired Date': 'First Hire Date', 'Time Spent Filling': 'Time to Fill', 'Started': 'Hires'}, inplace=True)
dfRequisition.head(5)
```

```python
## Adding yet another variable
dfRequisition['Time to Interview'] = dfRequisition['First Interview Date'] - dfRequisition['Date Created']

## Adding a variable to see how long they took to make a decision
dfRequisition['Decision Time'] = dfRequisition['First Hire Date'] - dfRequisition['First Interview Date']

dfRequisition.head(5)
```

```python
## For the time variables, fill the dates with a placeholder so that we can convert
dfRequisition = dfRequisition.fillna({'Time to Interview': 0, 'Time to Fill': 0, 'Decision Time': 0})
dfRequisition.head(5)
```

```python
## Change the variable types to int for the days
dfRequisition['Time to Interview'] = (dfRequisition['Time to Interview']/np.timedelta64(1, 'D')).astype(int)
dfRequisition['Time to Fill'] = (dfRequisition['Time to Fill']/np.timedelta64(1, 'D')).astype(int)
dfRequisition['Decision Time'] = (dfRequisition['Decision Time']/np.timedelta64(1, 'D')).astype(int)
dfRequisition.head(5)
```

```python
## Replace the 0s in the column with NaN
dfRequisition['Time to Interview'] = np.where(dfRequisition['Time to Interview'] ==0, np.NaN, dfRequisition['Time to Interview'])
dfRequisition['Time to Fill'] = np.where(dfRequisition['Time to Fill'] ==0, np.NaN, dfRequisition['Time to Fill'])
dfRequisition['Decision Time'] = np.where(dfRequisition['Decision Time'] ==0, np.NaN, dfRequisition['Decision Time'])
dfRequisition.head(5)
```

```python
dfRequisition.columns
```

```python
## Changing the order of the columns
dfRequisition = dfRequisition[['Requisition', 'Date Created', 'Company', 'Job Title', 'Salary', 'SAGE Fee', 'Status', 'First Interview Date', 
                               'Time to Interview', 'Decision Time','First Hire Date', 'Time to Fill', 'Pipelined', 'Submitted', 'Hires', 
                               'Web Apply', 'Phone Interview', 'Declined from Pipeline', 'In Person Interview', 'Video Interview', 
                               'Candidate Withdrew Interest', 'Client Declined', '2nd In Person Interview', 'Phone Screen', 'Candidate Declined', 
                               'Internal In Person Interview', 'Ready to Submit', 'Client Rejected Candidate', 'Declined from Web Apply', 
                               'Offer Declined', 'Offer', 'Applicants', 'HasPhoneInterview', 'HasInPersonInterview', 'HasVideoInterview', 
                               'Total Unique Candidates']]
dfRequisition.head(5)
```
